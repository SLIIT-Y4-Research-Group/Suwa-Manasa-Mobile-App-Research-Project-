```python
import glob

# Get all .txt files
txt_files = glob.glob("*.txt")
print(f"Found {len(txt_files)} story files.")

# Combine them into one large text
combined_text = ""
for file in txt_files:
    with open(file, 'r', encoding='utf-8') as f:
        combined_text += f.read() + "\n"

print(f"Total characters in dataset: {len(combined_text)}")
print("First 500 chars:\n", combined_text[:500])
```

```python
import re

# Remove unwanted characters (keep only Sinhala and basic punctuation)
def clean_text(text):
    text = re.sub(r'[^අ-ෆ.,!?\s]', '', text)  # Keep Sinhala, punctuation, spaces
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    return text.strip()

cleaned_text = clean_text(combined_text)
print(f"Cleaned text length: {len(cleaned_text)}")
print("Sample cleaned text:\n", cleaned_text[:300])
```

```python
# Get all unique characters
chars = sorted(list(set(cleaned_text)))
char_to_idx = {ch: i for i, ch in enumerate(chars)}
idx_to_char = {i: ch for i, ch in enumerate(chars)}

vocab_size = len(chars)
print(f"Vocabulary size: {vocab_size}")
print("First 10 chars in vocabulary:", chars[:10])

# Save mapping for later use (Flutter app)
import json
with open('char_to_idx.json', 'w', encoding='utf-8') as f:
    json.dump(char_to_idx, f, ensure_ascii=False)
```

```python
import numpy as np

# Convert text to indices
text_as_int = [char_to_idx[ch] for ch in cleaned_text]

# Prepare sequences for LSTM
seq_length = 100  # Input sequence length
examples_per_epoch = len(text_as_int) // (seq_length + 1)

dataX = []
dataY = []

for i in range(0, len(text_as_int) - seq_length):
    seq_in = text_as_int[i:i + seq_length]
    seq_out = text_as_int[i + seq_length]
    dataX.append(seq_in)
    dataY.append(seq_out)

X = np.array(dataX)
y = np.array(dataY)

print(f"Total training sequences: {len(X)}")
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout

# Remove input_length from Embedding (it's deprecated as the warning says)
model = Sequential([
    Embedding(vocab_size, 256),
    LSTM(256, return_sequences=True),
    Dropout(0.3),
    LSTM(256),
    Dropout(0.3),
    Dense(vocab_size, activation='softmax')
])

# Build the model with a sample batch
model.build(input_shape=(None, seq_length))  # (batch_size, sequence_length)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
# Train the model
epochs = 30
batch_size = 64

history = model.fit(X, y,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_split=0.2)

# Save the trained model
model.save('katha_mithura_lstm.h5')
print("Model saved as 'katha_mithura_lstm.h5'")
```

```python
# After training, history contains all metrics
print("\n Training History Metrics:")
print(f"Final Training Loss: {history.history['loss'][-1]:.4f}")
print(f"Final Training Accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"Final Validation Loss: {history.history['val_loss'][-1]:.4f}")
print(f"Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")

# Plot training progress
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
